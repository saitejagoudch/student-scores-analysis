### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
def grade(x):
    if x >= 90:
        return 'A'
    elif x < 90 and x >= 80:
        return 'B'
    elif x < 80 and x >= 70:
        return 'C'
    elif x < 70 and x >= 60:
        return 'D'
    elif x < 60:
        return 'F'

#apply method applies a function to each element of a DataFrame
df_students["CS 5000 Grade"] = df_students["CS 5000"].apply(lambda x:grade(x))
df_students
```

### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2022, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
def gpa(x):
    if (x >= 90):
        return 4.0
    if (x >= 80) & (x < 90):
        return 3.0
    if (x >= 70) & (x < 80):
        return 2.0
    if (x >= 60) & (x < 70):
        return 1.0
    return 0.0

#apply method applies a function to each element of a DataFrame
df_students['GPA'] = df_students.apply(lambda x: (((gpa(x['CS 5000']) * 3) +
                                                   (gpa(x['CS 5110']) * 3) +
                                                   (gpa(x['CS 5600']) * 3)) / 9), axis = 1)
df_students
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
+ `Normal: 3.0 <= GPA < 3.5`
+ `In Probation: GPA < 3.0`

```python
def Statusfun(x):
    if x >= 3.5:
        return "Honor Roll"
    elif x < 3.5 and x >= 3:
        return "Normal"
    elif x < 3:
        return "In Probation"

#apply method applies a function to each element of a DataFrame
df_students["Status"] = df_students["GPA"].apply(Statusfun)

df_students
```
